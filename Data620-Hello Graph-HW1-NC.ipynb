```python
#1. The code for Krackhardt graph, shows nodes as numbers
import matplotlib.pyplot as plt
import networkx as nx
G = nx.krackhardt_kite_graph()
nx.draw(G, with_labels=True)
plt.show()
```

```python
#2. Add the nodes: first try to add one
NG=nx.Graph()
NG.add_node('Andre') # show how to add a node 
nx.draw(NG,with_labels=True )
plt.show()
```

```python
NG.add_nodes_from(['Beverly','Ed','Carol','Diane','Fernando','Ed','Garth','Heather','Ike','Jane'])
nx.draw(NG,with_labels=True )
plt.show()
```

```python
# 3. Add the edges:
NG.add_edges_from([('Andre','Carol'),('Andre','Diane'),('Andre','Beverly'),('Fernando','Andre'), ('Beverly','Ed'),('Beverly','Diane'),('Beverly','Garth'),
                  ('Ed','Diane'),('Ed','Garth'),('Garth','Diane'),('Garth','Fernando'),('Garth','Heather'),
                  ('Heather','Fernando'),('Heather','Ike'),('Ike','Jane'),('Fernando','Diane'),('Fernando','Carol'),
                 ('Carol','Diane')])
plt.figure(figsize=(6,6))
nx.draw(NG,node_size=400,with_labels=True )
plt.title("Krackhardt Kite Graph", fontsize=15)
plt.show()
```
